Task 2: Dimensionality Reduction with PCA

1. Standardize the features: Use StandardScaler to standardize the features.
2. Implement PCA: Apply PCA to reduce the dimensionality of the dataset.
3. Determine the optimal number of principal components: Use a scree plot or cumulative explained variance to determine the optimal number of principal components.
4. Transform the original dataset: Transform the original dataset into the principal component

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

```

```python
# Standardize the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Implement PCA
pca = PCA()
pca_data = pca.fit_transform(df_scaled)

# Determine the optimal number of principal components
explained_variance = pca.explained_variance_ratio_
plt.plot(explained_variance)
plt.xlabel('Principal Component Index')
plt.ylabel('Explained Variance')
plt.show()
```

```python
# Select the top k principal components
k = 2  # Choose the number of principal components
pca = PCA(n_components=k)
pca_data = pca.fit_transform(df_scaled)


```

Task 3: Clustering with Original Data

1. Apply a clustering algorithm: Apply K-means clustering to the original dataset.
2. Visualize the clustering results: Use a scatter plot to visualize the clustering results.
3. Evaluate the clustering performance: Calculate the silhouette score or Davies-Bouldin index to evaluate the clustering performance.


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Apply K-means clustering
kmeans = KMeans(n_clusters=3)
kmeans.fit(df_scaled)
labels = kmeans.labels_

# Visualize the clustering results
plt.scatter(df_scaled[:, 0], df_scaled[:, 1], c=labels)
plt.show()

# Evaluate the clustering performance
silhouette = silhouette_score(df_scaled, labels)
print(f'Silhouette Score: {silhouette}')

```


Task 4: Clustering with PCA Data

1. Apply the same clustering algorithm: Apply K-means clustering to the PCA-transformed dataset.
2. Visualize the clustering results: Use a scatter plot to visualize the clustering results.
3. Compare the clustering results: Compare the clustering results obtained from the original dataset and PCA-transformed data.



```python
# Apply K-means clustering to PCA-transformed data
kmeans_pca = KMeans(n_clusters=3)
kmeans_pca.fit(pca_data)
labels_pca = kmeans_pca.labels_

# Visualize the clustering results
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=labels_pca)
plt.show()

# Compare the clustering results
silhouette_pca = silhouette_score(pca_data, labels_pca)
print(f'Silhouette Score (PCA): {silhouette_pca}')
```
