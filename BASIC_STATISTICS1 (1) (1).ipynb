Boxplot:
A boxplot, also called a box-and-whisker plot, is a graphical summary of numerical data that shows:

The spread of the data (how spread out it is),

The center of the data (median),

Outliers (values that are much higher or lower than the rest)

step 2:Load your csvfile

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("/content/sales_data_with_discounts.csv")
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_cols:
    plt.figure(figsize=(6, 1.2))
    sns.boxplot(x=df[col], color='green')
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.grid(True)
    plt.show()
```

To scale numerical values (like Sales, Discount, etc.) to have mean = 0 and standard deviation = 1. This makes the data uniform and improves performance for many machine learning algorithms.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/sales_data_with_discounts.csv')

# Select only numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Show original data (first 5 rows)
print("Original Data:\n", df[numerical_cols].head())

# Apply standardization using StandardScaler
scaler = StandardScaler()
df_standardized = df.copy()
df_standardized[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Show standardized data
print("\nStandardized Data:\n", df_standardized[numerical_cols].head())
# Plot comparison for a specific column (e.g., 'Sales')
plt.figure(figsize=(12, 5))

# Original
plt.subplot(1, 2, 1)
plt.hist(df['Total Sales Value'], bins=20, color='skyblue', edgecolor='black')
plt.title("Original Sales Distribution")

# Standardized
plt.subplot(1, 2, 2)
plt.hist(df_standardized['Total Sales Value'], bins=20, color='orange', edgecolor='black')
plt.title("Standardized Sales Distribution")

plt.tight_layout()
plt.show()
```
