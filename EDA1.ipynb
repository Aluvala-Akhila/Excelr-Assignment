```python
# detecting outliers using IQR
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR= Q3 - Q1
df_out = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(f"Original dataset shape: {df.shape}")
print(f"After removing outliers: {df_out.shape}")
```

```python
# 3.Data Visulization
#1. Boxplot
#2. Histogram
# import required librabries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Set plot style
sns.set(style="whitegrid")
# Create a figure and plot boxplots
plt.figure(figsize=(15, 8))
sns.boxplot(data=df, palette="pastel")
plt.title("Boxplot of All Numerical Features in Cardiographic Dataset", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# pie chart to display the frequency of categories for  categorical variables.
import matplotlib.pyplot as plt
import pandas as pd

# Use the existing DataFrame 'df' instead of reloading
# Check column name â€” it must match exactly
print(df.columns)

# Count unique values in NSP
category_counts = df['NSP'].value_counts()

# Prepare pie chart inputs
labels = category_counts.index.map({1.0: "Normal", 2.0: "Suspect", 3.0: "Pathological"})
sizes = category_counts.values
colors = ['lightgreen', 'gold', 'salmon']

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title("Fetal Health Category Distribution")
plt.axis('equal')  # Ensures pie is a circle
plt.show()
```
